# Augmented-Dicky-Fuller test


## test_statastics < critical_values
## p-value < 0.05

```python
from statsmodels.tsa.stattools import adfuller
```

```python
adfTest=adfuller(data['passengers'],autolag='AIC')
```

```python
adfTest
```

### not stationary data

```python
def adf_test(timeseries):
    print('Result of dickyFuller test :- ')
    dftest=adfuller(timeseries , autolag='AIC')
    dfoutput= pd.Series(dftest[0:4], index=['Test statistics','p-value','lags-used','Number of observation used'])
    for key, value in dftest[4].items():
        dfoutput['Critical Value (%s)'%key] = value
    print(dfoutput)   
```

```python
adf_test(data['passengers'])
```

```python
def stationarity_test(dataframe, var):
    dataframe['rollMean']=dataframe[var].rolling(window=12).mean()
    dataframe['rollStd']=dataframe[var].rolling(window=12).std()
    
    print('Result of dickyFuller test :- ')
    dftest=adfuller(dataframe[var] , autolag='AIC')
    dfoutput= pd.Series(dftest[0:4], index=['Test statistics','p-value','lags-used','Number of observation used'])
    print(dfoutput)
    
    for key, values in dftest[4].items():
        print('criticality', key ,' : ',values)
     
    
    sns.lineplot(data=dataframe, x=dataframe.index, y=var,label=var)
    sns.lineplot(data=dataframe, x=dataframe.index, y='rollMean',label='rollMean')
    sns.lineplot(data=dataframe, x=dataframe.index, y='rollStd' ,label='rollStd')
    plt.legend()
    plt.show()
```

```python
airp=data[['passengers']].copy(deep=True)
airp['diff_1']=airp['passengers'].diff()
airp['diff_12']=airp['passengers'].diff(12)
airp.head()
```

```python
from statsmodels.tsa.arima.model import ARIMA #
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf
```

```python
# p(AR)
# d(I)- = diff(n), d=n
# q(AM)
```

```python
plot_pacf(airp['diff_1'].dropna(),lags=20);
#shaded_area is call insignificant area and out_side area is calles significant area
#(significant_area)p=i
```

```python
plot_acf(airp['diff_1'].dropna(),lags=20);
# q=3 as d=1 (dif_1)
```

```python
train=airp[:round(len(airp)*0.7)]

train.head()
#train.shape
```
